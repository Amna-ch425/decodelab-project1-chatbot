## 📌 What is a Rule-Based Chatbot?
A **rule-based chatbot** works by matching user input to predefined patterns and returning fixed responses. It uses `if-else` logic — no machine learning involved. It's the simplest form of AI-driven conversation.

## Step 1: Define the Response Rules

```python
# ============================================================
# RULE-BASED CHATBOT — DecodeLab Internship Project 1
# ============================================================

def get_response(user_input):
    """
    Takes user input (string), returns a chatbot response (string).
    Uses if-else logic to match keywords and return predefined answers.
    """
    # Convert to lowercase for easier matching
    text = user_input.lower().strip()

    # ---- GREETINGS ----
    if text in ["hi", "hello", "hey", "good morning", "good evening", "howdy"]:
        return "👋 Hello! I'm DecodeBot. How can I help you today?"

    # ---- HOW ARE YOU ----
    elif "how are you" in text or "how r u" in text:
        return "😊 I'm just a bot, but I'm doing great! What about you?"

    # ---- NAME ----
    elif "your name" in text or "who are you" in text:
        return "🤖 I'm DecodeBot, your AI assistant built during the DecodeLab Internship!"

    # ---- WHAT CAN YOU DO ----
    elif "what can you do" in text or "help" in text or "commands" in text:
        return (
            "📋 I can help you with the following topics:\n"
            "  - Greetings (hi, hello, hey)\n"
            "  - Weather info\n"
            "  - Jokes\n"
            "  - Time & Date\n"
            "  - Basic math\n"
            "  Type 'bye' or 'exit' to quit."
        )

    # ---- WEATHER ----
    elif "weather" in text:
        return "🌤️ I can't check live weather yet, but it's always a great day to code!"

    # ---- JOKE ----
    elif "joke" in text or "funny" in text:
        return "😂 Why do programmers prefer dark mode? Because light attracts bugs! 🐛"

    # ---- TIME ----
    elif "time" in text:
        from datetime import datetime
        now = datetime.now().strftime("%H:%M:%S")
        return f"🕐 Current time is: {now}"

    # ---- DATE ----
    elif "date" in text or "today" in text:
        from datetime import datetime
        today = datetime.now().strftime("%A, %B %d, %Y")
        return f"📅 Today is: {today}"

    # ---- BASIC MATH ----
    elif "calculate" in text or "math" in text or any(op in text for op in ["+", "-", "*", "/"]):
        try:
            # Extract and evaluate a safe math expression
            import re
            expr = re.findall(r"[\d\+\-\*\/\.\(\) ]+", text)
            if expr:
                result = eval(expr[0].strip())
                return f"🔢 Result: {result}"
        except:
            pass
        return "🔢 Please type a math expression like: calculate 5 + 3"

    # ---- INTERNSHIP ----
    elif "internship" in text or "decodelab" in text:
        return "🎓 DecodeLab Internship is helping you build real AI projects step by step. Keep going!"

    # ---- THANKS ----
    elif "thank" in text or "thanks" in text:
        return "😊 You're welcome! Happy to help."

    # ---- EXIT COMMANDS ----
    elif text in ["bye", "exit", "quit", "goodbye", "see you"]:
        return "👋 Goodbye! Happy coding! Come back anytime. 🚀"

    # ---- DEFAULT (unknown input) ----
    else:
        return "🤔 Sorry, I didn't understand that. Type 'help' to see what I can do!"


print("✅ Chatbot function defined successfully!")
```

## Step 2: Test Individual Responses

```python
# Test a few responses manually
test_inputs = [
    "hello",
    "what is your name?",
    "tell me a joke",
    "what's the time?",
    "calculate 15 + 27",
    "bye"
]

print("=" * 50)
print("   CHATBOT TEST RESPONSES")
print("=" * 50)
for inp in test_inputs:
    print(f"\n👤 You: {inp}")
    print(f"🤖 Bot: {get_response(inp)}")
```

## Step 3: Run the Chatbot in a Loop (Interactive Mode)

```python
# ============================================================
#  INTERACTIVE CHATBOT LOOP
#  Type messages and the bot will respond.
#  Type 'bye' or 'exit' to quit.
# ============================================================

print("=" * 50)
print("   Welcome to DecodeBot! 🤖")
print("   Type 'help' to see commands.")
print("   Type 'bye' to exit.")
print("=" * 50)

while True:
    user_input = input("\n👤 You: ")

    if not user_input.strip():
        print("🤖 Bot: Please type something!")
        continue

    response = get_response(user_input)
    print(f"🤖 Bot: {response}")

    # Stop the loop on exit commands
    if user_input.lower().strip() in ["bye", "exit", "quit", "goodbye", "see you"]:
        break
```
